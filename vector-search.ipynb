```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd

books = pd.read_csv('books_cleaned.csv')
```

```python
books
```

```python
books["tagged_description"].to_csv("tagged_description.txt", sep="\n", index=False, header=False)
```

```python
raw_document = TextLoader("tagged_description.txt", encoding="utf-8").load()
text_splitter = CharacterTextSplitter(chunk_size=10, chunk_overlap=0, separator="\n")
document = text_splitter.split_documents(raw_document)
```

```python
document[0]
```

```python
db_books = Chroma.from_documents(
    document,
    embedding=OpenAIEmbeddings())
```

```python
query ="A book about war"
docs = db_books.similarity_search(query,k=10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]
```

```python
def semantic_recommendations(query: str, k:int=50) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k=k)
    books_list = []
    
    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]
        
    return books[books["isbn13"].isin(books_list)]
    
```

```python
semantic_recommendations("A book about war")
```
